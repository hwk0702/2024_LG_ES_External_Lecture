# Import modules

```python
import numpy as np
import pandas as pd
import random
import os

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
from torch.utils.data import Dataset, DataLoader

from torchvision import datasets
from torchvision import transforms

import matplotlib.pyplot as plt
import seaborn as sns

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
```

# Functions

```python
def torch_seed(random_seed):
    torch.manual_seed(random_seed)
    torch.cuda.manual_seed(random_seed)
    torch.cuda.manual_seed_all(random_seed) # if use multi-GPU
    # CUDA randomness
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    np.random.seed(random_seed)
    random.seed(random_seed)
    os.environ['PYTHONHASHSEED'] = str(random_seed)


def train(
    model, dataloader, criterion, optimizer,
    log_interval: int, device: str, add_noise: bool = False) -> list:

    total_loss = []

    model.train()

    for i, (images, _) in enumerate(dataloader):
        # convert device
        images = images.to(device)

        if add_noise and (np.random.random(1) > 0.5):
            # add noise
            noise_factor = 0.5
            inputs = torch.clamp(
                images + (noise_factor * torch.randn(*images.shape)).to(device),
                min = 0,
                max = 1
            )
        else:
            inputs = images

        # model outputs
        outputs = model(inputs)

        # loss
        loss = criterion(images, outputs)
        total_loss.append(loss.item())

        # calculate gradients
        loss.backward()

        # update model weights
        optimizer.step()
        optimizer.zero_grad()

        # log learning history
        if i % log_interval == 0 or (i+1) == len(dataloader):
            print(f"{'TRAIN':5s} [{i+1:5d}/{len(dataloader):5d}] loss: {np.mean(total_loss):.4f}")

    # average loss
    avg_loss = np.mean(total_loss)

    return avg_loss

def test(
    model, dataloader, criterion,
    log_interval: int, device: str, add_noise: bool = False, return_latent: bool = False) -> list:

    total_loss = []
    latents = []

    torch_seed(223)
    model.eval()

    with torch.no_grad():
        for i, (images, _) in enumerate(dataloader):
            # convert device
            images = images.to(device)

            if add_noise and (np.random.random(1) > 0.5):
                # add noise
                noise_factor = 0.5
                inputs = torch.clamp(
                  images + (noise_factor * torch.randn(*images.shape)).to(device),
                  min = 0,
                  max = 1
                )
            else:
                inputs = images

            # latent outputs
            if return_latent:
                latents.append(model.encoder(inputs).cpu().numpy())
            else:
                # model outputs
                outputs = model(inputs)

                # loss
                loss = criterion(images, outputs)
                total_loss.append(loss.item())

                # log learning history
                if i % log_interval == 0 or (i+1) == len(dataloader):
                    print(f"{'TSET':5s} [{i+1:5d}/{len(dataloader):5d}] loss: {np.mean(total_loss):.4f}")

    # average loss
    avg_loss = np.mean(total_loss)

    # return
    if return_latent:
        return np.vstack(latents)
    else:
        return avg_loss


def fit(
    model, trainloader, testloader, criterion, optimizer,
    epochs: int, log_interval: int, device: str, add_noise: bool = False) -> list:

    train_history = []
    test_history = []

    # fitting model
    for i in range(epochs):
        print(f'\nEpoch: [{i+1}/{epochs}]')
        train_loss = train(
            model        = model,
            dataloader   = trainloader,
            criterion    = criterion,
            optimizer    = optimizer,
            log_interval = log_interval,
            device       = device,
            add_noise    = add_noise
        )

        test_loss = test(
            model        = model,
            dataloader   = testloader,
            criterion    = criterion,
            log_interval = log_interval,
            device       = device
        )

        print(f'\nTest loss: {test_loss:.4f}')

        # show results
        with torch.no_grad():
            fig, ax = plt.subplots(2, 10, figsize=(25,5))

            for i in range(10):
                idx = torch.where(testloader.dataset.targets==i)[0][0]
                img = testloader.dataset[idx][0]
                output = model(img.unsqueeze(0).to(device)).cpu()[0]
                ax[0, i].imshow(img.permute(1,2,0), cmap='gray')
                ax[1, i].imshow(output.permute(1,2,0), cmap='gray')
            plt.tight_layout()
            plt.show()

        # stack history
        train_history.append(train_loss)
        test_history.append(test_loss)

    return train_history, test_history


def figure(all_train_history: list, all_test_history: list, all_exp_name: list) -> None:
    fig, ax = plt.subplots(1, 2, figsize=(12,5))

    # train line plot
    for i, (train_h, exp_name) in enumerate(zip(all_train_history, all_exp_name)):
        sns.lineplot(
            x     = range(1, len(train_h)+1),
            y     = train_h,
            label = exp_name,
            ax    = ax[0]
        )

    # test lineplot
    for i, (test_h, exp_name) in enumerate(zip(all_test_history, all_exp_name)):
        sns.lineplot(
            x     = range(1, len(test_h)+1),
            y     = test_h,
            label = exp_name,
            ax    = ax[1]
        )

    # set y axis label
    ax[0].set_ylabel('MSE Loss')
    ax[1].set_ylabel('MSE Loss')

    # set x axis label
    ax[0].set_xlabel('Epochs')
    ax[1].set_xlabel('Epochs')

    # set title
    ax[0].set_title('Train history')
    ax[1].set_title('Test history')

    # set y value limit
    max_train = np.max(all_train_history)
    max_test = np.max(all_test_history)
    ax[0].set_ylim(0, max_train+0.01)
    ax[1].set_ylim(0, max_test+0.01)

    # set legend
    ax[0].legend(loc='upper right')
    ax[1].legend(loc='upper right')
    plt.tight_layout()
    plt.show()
```

# Auto-Encoder

## Linear Auto-Encoder

```python
class AutoEncoder(nn.Module):
    def __init__(self, input_dim: int, dims: list):
        super().__init__()

        dims = [input_dim] + dims

        self.flatten = nn.Flatten()
        self.enc = nn.Sequential(*self.build_layer(dims=dims))
        self.dec = nn.Sequential(*self.build_layer(dims=dims[::-1]))
        self.output = nn.Linear(in_features=input_dim, out_features=input_dim)

    def build_layer(self, dims):
        layer = []
        for i in range(1, len(dims)):
            layer_i = [
                nn.Linear(in_features=dims[i-1], out_features=dims[i]),
                nn.ReLU()
            ]
            layer.extend(layer_i)

        return layer


    def encoder(self, x):
        out = self.flatten(x)
        out = self.enc(out)

        return out

    def decoder(self, out, x_shape):
        out = self.dec(out)
        out = self.output(out)
        out = F.sigmoid(out)
        out = out.view(x_shape)

        return out

    def forward(self, x):
        out = self.encoder(x)
        out = self.decoder(out, x.shape)

        return out
```

```python
torch_seed(cfg.seed)
ae = AutoEncoder(input_dim = np.prod(cfg.image_size), dims = [128, 32])
ae.to(cfg.device)
print('load Auto-Encoder')
print('The number of model parameters: ',sum([p.numel() for p in ae.parameters()]))

criterion = nn.MSELoss()
optimizer = SGD(ae.parameters(), lr=cfg.learning_rate)
```

## Convolutional Auto-Encoder

```python
class ConvolutionalAutoEncoder(nn.Module):
    def __init__(self, input_dim: int, dims: list):
        super().__init__()

        dims = [input_dim] + dims

        self.enc = nn.Sequential(*self.build_layer(dims=dims))
        self.dec = nn.Sequential(*self.build_layer(dims=dims[::-1], up=True))
        self.output = nn.Conv2d(
            in_channels  = input_dim,
            out_channels = input_dim,
            kernel_size  = 3,
            padding      = 1
        )

    def build_layer(self, dims, up=False):
        layer = []

        for i in range(1, len(dims)):
            if up:
                layer_i = [
                    nn.ConvTranspose2d(
                        in_channels  = dims[i-1],
                        out_channels = dims[i],
                        kernel_size  = 2,
                        stride       = 2
                    ),
                    nn.ReLU()
                ]
            else:
                layer_i = [
                    nn.Conv2d(
                        in_channels  = dims[i-1],
                        out_channels = dims[i],
                        kernel_size  = 3,
                        padding      = 1),
                    nn.ReLU(),
                    nn.MaxPool2d(kernel_size=2, stride=2)
                ]

            layer.extend(layer_i)

        return layer

    def encoder(self, x):
        out = self.enc(x)

        return out

    def decoder(self, out):
        out = self.dec(out)
        out = self.output(out)
        out = F.sigmoid(out)

        return out

    def forward(self, x):
        out = self.encoder(x)
        out = self.decoder(out)

        return out
```

```python
torch_seed(cfg.seed)
cae = ConvolutionalAutoEncoder(input_dim=1, dims=[16,32])
cae.to(cfg.device)
print('load Convolutional Auto-Encoder')
print('The number of model parameters: ',sum([p.numel() for p in cae.parameters()]))

criterion = nn.MSELoss()
optimizer = SGD(cae.parameters(), lr=cfg.learning_rate)
```

## Visualization image reconstruction outputs

```python
fig, ax = plt.subplots(10, 10, figsize=(25,25))

column_list = ['image'] + list(models.keys()) + ['noise'] + list(models.keys())

for col, name in enumerate(column_list):
    for row in range(10):
        idx = torch.where(testset.targets==row)[0][0]
        img = testset[idx][0]

        # set image for visulization
        if col < column_list.index('noise'):
            img = img
        else:
            torch_seed(cfg.seed)
            img = img + torch.randn(*img.size()) * 0.5

        # get model output
        if (name != 'image') and (name != 'noise'):
            with torch.no_grad():
                img = models[name](img.unsqueeze(0).to(cfg.device)).cpu()[0]

        ax[row, col].imshow(img.permute(1,2,0), cmap='gray')
        ax[row, col].set_title(name)
        ax[row, col].axis('off')

plt.tight_layout()
plt.show()
```

## Visualization feature embedding space using T-SNE

```python
latents_wo_noise = []
latents_w_noise = []

for name, model in models.items():
    print(f'\n[{name}]')
    print('Test images without noise')
    latents_wo_noise.append(test(
        model         = model,
        dataloader    = testloader,
        criterion     = criterion,
        log_interval  = cfg.log_interval,
        device        = cfg.device,
        return_latent = True
    ))
    print('Test images with noise')
    latents_w_noise.append(test(
        model         = model,
        dataloader    = testloader,
        criterion     = criterion,
        log_interval  = cfg.log_interval,
        device        = cfg.device,
        add_noise     = True,
        return_latent = True
    ))
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, init='random')
```

```python
select_idx = np.random.choice(range(len(testset)), size=2000, replace=False)

fig, ax = plt.subplots(2, len(models), figsize=(20,10), constrained_layout=False)

for col_idx, name in enumerate(models.keys()):
    # T-SNE without noise
    X_embed = tsne.fit_transform(latents_wo_noise[col_idx].reshape(len(testset), -1)[select_idx])
    sns.scatterplot(
        x         = X_embed[:,0],
        y         = X_embed[:,1],
        hue       = testset.targets.numpy().astype(str)[select_idx],
        hue_order = np.array(list(testset.class_to_idx.values())).astype(str),
        ax        = ax[0, col_idx]
    )

    # T-SNE with noise
    X_embed = tsne.fit_transform(latents_w_noise[col_idx].reshape(len(testset), -1)[select_idx])
    sns.scatterplot(
        x         = X_embed[:,0],
        y         = X_embed[:,1],
        hue       = testset.targets.numpy().astype(str)[select_idx],
        hue_order = np.array(list(testset.class_to_idx.values())).astype(str),
        ax        = ax[1, col_idx]
    )

    # set title
    ax[0, col_idx].set_title(f'{name}\nwithout noise')
    ax[1, col_idx].set_title(f'{name}\nwith noise')

    # set legend
    ax[0, col_idx].legend(loc='upper right')
    ax[1, col_idx].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
